```python
#mudando o objeto date para datetime64
df['date']= pd.to_datetime(df['date']) 
df.dtypes
```

```python
df.columns = ['Data', 'UF', 'Município', 'Tipo_de_local', 'Casos_Confirmados', 'Mortes', 'Dados_Atualizados',
       'População_estimada_2019', 'Código IBGE',
       'Confirmados_por_100k_hab', 'Mortes/confirmados']
df.sample(5)
```

```python
df1 = df.loc[(df['Dados_Atualizados'] == True) & (df['Tipo_de_local'] == 'state') ]
df1.sample(3)
```

```python
df1.shape
```

```python
# NÚMERO DE MORTES TOTAL (INCLUINDO SUSPEITAS E CASOS CONFIRMADOS DE CORONAVIRUS)
df_uf_mortes = df1.groupby('UF').Mortes.sum()
df_uf_mortes.sort_values(ascending=False)
```

```python
# NÚMERO DE CASOS CONFIRMADOS DE CORONAVIRUS 
df_uf_cases = df1.groupby('UF').Casos_Confirmados.sum()
df_uf_cases.sort_values(ascending=False)
```

```python
#número de casos por 100 mil habitantes por UF
df_uf_pop = df1.groupby('UF').População_estimada_2019.sum()
casos_por_100k_habitantes = (df_uf_cases/df_uf_pop)*100000
casos_por_100k_habitantes.sort_values(ascending=False)
```

```python
# Evolução na cidade de São Paulo
df_sp_evo = df.loc[(df['UF'] == 'SP') & (df['Tipo_de_local'] == "state")]
df_sp_evo = df_sp_evo.reset_index()
df_sp_evo = df_sp_evo.loc[df_sp_evo['Data'] >= '2020-03-16']
df_sp_evo.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# define size of figure
width_in_inches = 15
height_in_inches = 7.5
dots_per_inch = 70

plt.figure(
    figsize=(width_in_inches, height_in_inches),
    dpi=dots_per_inch)

Data = df_sp_evo['Data']
Mortes = df_sp_evo['Mortes']

# Plot the bar graph
plot = plt.bar(Data, Mortes)

# Add the data value on head of the bar
for value in plot:
    height = value.get_height()
    plt.text(value.get_x() + value.get_width()/2.,
             1.002*height,'%d' % int(height), ha='center', va='bottom')

# Add labels and title
plt.title('Evolução das mortes por COVID-19 no Estado de São Paulo')
plt.xlabel('Data')
plt.ylabel('Número de mortos')

plt.tight_layout()
plt.savefig("out.png")

plt.show()
```
